### Access Tokens

```python
import requests
import json
from pandas.io.json import json_normalize
import pandas as pd

client_id = ''
api_key = ''
```

### Request

```python
def get_data_yelpAPI(term, locations, api_key):
    
    headers = {
            'Authorization': 'Bearer {}'.format(api_key),
        }
    url = 'https://api.yelp.com/v3/businesses/search'
    
    raw_data = pd.DataFrame()
    
    for l in locations:

        for offset in range(0, 1000, 50):

            url_params = {
                'limit': 50,
                'term': term.replace(' ', '+'),
                'location': l.replace(' ', '+'),
                'offset': offset
            }

            response = requests.get(
                url,
                headers = headers,
                params = url_params
            )

            if response.status_code == 200:
                data = json.loads(response.text)['businesses']
                data = json_normalize(data)
                raw_data = raw_data.append(data)

            elif response.status_code == 400:
                print('400 Bad Request')
                break
                
    # Write out
    upload_dir = ''
    upload_file = 'raw_yelp_data'
    raw_data.to_csv(upload_dir + upload_file + '.csv', index = False)
```

### DataFrame Cleaning

```python
raw_data = pd.read_csv('')

gym_data = raw_data[[
    'name',
    'location.city',
    'location.state',
    'coordinates.latitude',
    'coordinates.longitude'
]]

gym_data = gym_data.rename(columns = {
    'location.city': 'city',
    'location.state': 'state',
    'coordinates.latitude': 'latitude',
    'coordinates.longitude': 'longitude'
})

gym_data['latitude'] = gym_data.latitude.apply(lambda x: round(x, 4))
gym_data['longitude'] = gym_data.longitude.apply(lambda x: round(x, 4))
gym_data = gym_data.dropna().reset_index(drop = True)

# Write out
upload_dir = ''
upload_file = 'clean_yelp_data'
gym_data.to_csv(upload_dir + upload_file + '.csv', index = False)
```
